## Amazon

```python
# Importing imp libs
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

```python
#importing the data
df = pd.read_csv("C:\\Users\\sredekar\\OneDrive - Interpublic\\Desktop\\PW Skills\\EDA 2\\amazon.csv",  sep=',')

print(f'No of rows in the data {df.shape}')
df.head()
```

```python
# Data info

df.info()
```

```python
## Preprocessing 

df['discounted_price'] = pd.to_numeric(df['discounted_price'].str.replace('₹', '').str.replace(',', ''))
df['actual_price'] =  pd.to_numeric(df['actual_price'].str.replace('₹', '').str.replace(',', ''))
df['rating'] = pd.to_numeric(df['rating'].str.replace('|',''))
df['rating_count'] = pd.to_numeric(df['rating_count'].str.replace(',',''))
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.replace('%',''))
df['Product_clean'] = [i.strip().split()[0] for i in df.product_name]

df["category_final"] = [i.split('|')[0] for i in df['category']]
df["category_final"].value_counts(normalize = True)*100

```

```python
df.info()
```

```python
df.describe()
```

```python
# What is the average rating for each product category

rating_per_category = df.groupby('category_final')['rating'].mean()

plt.figure(figsize=(10, 6))
rating_per_category.plot(kind='bar', color='red', alpha=0.7)

plt.title('Avg rating per category')
plt.xlabel('Category')
plt.ylabel('Avg rating')

plt.show()
```

```python
#What are the top rating_count products by category ?

avg_rating_per_product_per_cat = df.groupby(['category_final','product_name'])['rating'].mean().reset_index()
idmax = avg_rating_per_product_per_cat.groupby(['category_final'])['rating'].idxmax()
top_rated_product_per_cat = avg_rating_per_product_per_cat.loc[idmax]
top_rated_product_per_cat
```

```python
# What is the distribution of discounted prices vs. actual prices?

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='discounted_price', y='actual_price', data=df, color='blue')

# Adding title and labels
plt.title('discounted_price vs actual_price of products')
plt.xlabel('discounted_price')
plt.ylabel('actual_price')

# Display the plot
plt.show()
```

Relationship between actual_price and discounted_price is linear

```python
# How does the average discount percentage vary across categories?

avg_dis_per_cat = df.groupby('category_final')['discount_percentage'].mean()

plt.figure(figsize=(8, 5))
avg_dis_per_cat.plot(kind='bar', color=['green'], alpha=0.7)

plt.title('Avg Discount_percentage by catgeories')
plt.xlabel('Categories')
plt.ylabel('discount_percentage')
plt.show()

print(avg_dis_per_cat)
```

%Discount is highest for HomeImprovement & lowest for OfficeProducts

```python
# 5. What are the most popular product names?

top_20_products = df['Product_clean'].value_counts(normalize = True).head(20)*100

plt.figure(figsize=(8, 5))
top_20_products.plot(kind='bar', color=['green'], alpha=0.7)

plt.title('Top 20 products')
plt.xlabel('Products')
plt.ylabel('% of Total')
plt.show()

print(top_20_products)
```

```python
# 6. What are the most popular product reviews?
from collections import Counter

df['clean_reviews'] = df['review_title'].str.split(',')

all_reviews = [item.strip() for sublist in df['clean_reviews'] for item in sublist]

review_counter = Counter(all_reviews)

most_common_reviews = review_counter.most_common(10)

print("Most popular reviews:")
for review, count in most_common_reviews:
    print(f"'{review}': {count} times")
```

```python
# 6. What are the most popular product keywords?
df['clean_product_info'] = df['about_product'].str.upper().str.split(':')

all_reviews = [item.strip() for sublist in df['clean_product_info'] for item in sublist]

review_counter = Counter(all_reviews)

most_common_reviews = review_counter.most_common(10)

print("Most popular product keywords:")
for review, count in most_common_reviews:
    print(f"'{review}': {count} times")
```

```python
# 8. What is the correlation between discounted_price and rating?

plt.figure(figsize=(10, 6))
sns.scatterplot(x='discounted_price', y='rating', data=df, color='blue')

plt.title('Scatter Plot of Discounted_price vs rating')
plt.xlabel('Discounted_price')
plt.ylabel('Rating')
plt.show()
```

High Discounted products get higher ratings 

```python
# 9. What are the Top 5 categories based on the highest ratings?

avg_rating_per_cat = df.groupby('category_final')['rating'].mean()

plt.figure(figsize=(8, 5))
avg_rating_per_cat_sorted = avg_rating_per_cat.sort_values(ascending=False)
avg_rating_per_cat_sorted.plot(kind='bar', color=['green'], alpha=0.7)

plt.title('Avg rating by categories')
plt.xlabel('Categories')
plt.ylabel('Rating')
plt.show()

print(avg_rating_per_cat_sorted.head())
```

```python
# 10. Identify any potential areas for improvement or optimization based on the data analysis.

# Lets plot products with lowest avg ratings

avg_rating_per_cat = df.groupby('Product_clean')['rating'].mean()

plt.figure(figsize=(8, 5))
avg_rating_per_cat_sorted = avg_rating_per_cat.sort_values(ascending=True).head(10)
avg_rating_per_cat_sorted.plot(kind='bar', color=['green'], alpha=0.7)

plt.title('Avg rating by Product')
plt.xlabel('Product')
plt.ylabel('Rating')
plt.show()

print(avg_rating_per_cat_sorted.head(10))
```

```python
#importing the data
df = pd.read_csv("C:\\Users\\sredekar\\OneDrive - Interpublic\\Desktop\\PW Skills\\EDA 2\\spotify.csv",  sep=',')

print(f'No of rows in the data {df.shape}')
df.head()
```

```python
#.Load the dataframe and ensure data quality by checking for missing values and duplicate rows.
     ##Handle missing values and remove duplicate rows if necessary.

df.isnull().sum()

# We can see there are no missing values
```

```python

print(f'Rows before dropping the duplicates {df.shape}')

print(f'Rows before dropping the duplicates {df.drop_duplicates().shape}')

df = df.drop_duplicates()
```

```python
df.describe()
```

```python
# 2.What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.

plt.figure(figsize=(10, 6))
df['Popularity'].plot(kind='hist', color='red', alpha=0.7)

plt.title('Popularity distribution')
plt.xlabel('Popularity')
plt.show()
```

The Popularity is normally distrubuted with right skew

```python
# 3.Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Popularity', y='Duration (ms)', data=df, color='blue')

plt.title('Scatter Plot of Popularity vs Duration (ms)')
plt.xlabel('Popularity')
plt.ylabel('Duration (ms)')
plt.show()
```

Tracks with lower duration has higher popularity

```python
df.columns
```

```python
# 4.Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist

track_counts = df['Artist'].value_counts()
top_10_artists = track_counts.head(10).index
df_top_10 = df[df['Artist'].isin(top_10_artists)]
plt.figure(figsize=(10, 6))
sns.countplot(data=df_top_10, x='Artist', palette='viridis', order=top_10_artists)
plt.title('Number of Tracks per Artist (Top 10)', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Count of Tracks', fontsize=14)
plt.show()
```

```python
# 5.What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.

df.groupby(['Track Name','Artist'])['Popularity'].sum().reset_index().sort_values(by = 'Popularity', ascending = True).head()
```

```python
# Among the top 5 most popular artists, which artist has the highest popularity on average?
    ## Calculate and display the average popularity for each artist.

avg_pop_artist = df.groupby('Artist')['Popularity'].mean().reset_index().sort_values(by = 'Popularity', ascending = False).head()

plt.figure(figsize=(8, 5))
plt.bar(avg_pop_artist['Artist'], avg_pop_artist['Popularity'], color='green', alpha=0.7)

plt.title('Avg Popularity by Artist')
plt.xlabel('Artist')
plt.ylabel('Avg Popularity')
plt.show()

print(avg_pop_artist)
```

```python
# 7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.

mask = df.groupby(['Artist','Track Name'])['Popularity'].sum().reset_index().sort_values(by = ['Artist','Popularity'], ascending = [True,False])

mask = mask.groupby(['Artist']).first().reset_index()

mask.merge(right = avg_pop_artist['Artist'], on = 'Artist', how ='inner')  
                                                                                         
```

```python
# 8.Visualize relationships between multiple numerical variables simultaneously using a pair plot.

data = pd.DataFrame(df[['Popularity','Duration (ms)']])

# Create a pair plot
sns.pairplot(data, height=3, aspect=1.2, markers='o', diag_kind='kde')

# Add title
plt.suptitle('Pair Plot of Popularity and Duration', y=1.02, fontsize=16)

# Display the plot
plt.show()
```

```python
# 9.Does the duration of tracks vary significantly across different artists?
        ## Explore this visually using a box plot or violin plot

## plotting it for top 5 artists

random_artists = np.random.choice(df['Artist'], size=10, replace=False).tolist()
df_artist = df[df['Artist'].isin(random_artists)]

plt.figure(figsize=(10, 6))
sns.violinplot(x='Artist', y='Duration (ms)', data=df_artist, palette='viridis')
plt.title('Distribution of Track Durations Across Artists', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Duration (minutes)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
