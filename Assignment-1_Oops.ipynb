Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
import logging as log
                  
log.basicConfig(filename = 'problem1.log',level = log.DEBUG,format = '%(asctime)s- %(message)s - %(levelname)s ')

class Bank_account():
    
    def __init__(self, account_number , Name , balance):
        self.account_number = account_number
        self.Name = Name
        self.balance = balance
        
    def deposit_money(self, deposit):       
        try:  
             if not isinstance(deposit, int) or deposit <= 0:              
                print('You have entered invalid value')
                log.error('You have entered invalid value')
             
             else:
                    self.balance += deposit
                    print(f'You have deposited ${deposit} in the account & balance is ${self.balance}')
                    log.info(f'Deposited ${deposit} into the account & balance is ${self.balance}')
        
        except Exception as e:
            log.error(e)
                      
                      
    def withdraw_money(self, withdraw):
        try:
            if withdraw <= self.balance:
                self.balance -= withdraw
                print(f'You have withdrawn ${withdraw} from the account')
                log.info(f'Withdrawn ${withdraw} from the account & balance is ${self.balance}')
                
            else:
                print(f'You do not have sufficient funds to withdraw')
                log.error('Amount entered is more than the balance')
                
        except Exception as e:
            log.error(e)
                
    def display_balance(self):
        print(f'Account Balance for {self.Name}: ${self.balance}')
        
acct1=Bank_account(123,'Shiv',200)

acct1.deposit_money(300)
acct1.display_balance()
acct1.deposit_money(-40)
acct1.display_balance()
acct1.withdraw_money(300)
acct1.display_balance()
acct1.withdraw_money(300)
acct1.display_balance()        
```

Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

```python
import logging as log

log.basicConfig(filename = 'problem3.log',level = log.DEBUG,format = '%(asctime)s- %(message)s - %(levelname)s ')

class Vehicle_rent():
    
    total_cars = 10
                    
    def Available_cars(self):
        if self.total_cars == 0:
            print('We do not have any cars to rent')
        else:
            print(f'Total cars available for renting are {self.total_cars}')
        
    def rent(self, cars):       
        try:
            
            if not isinstance(cars, int) or cars < 0  :              
                print('You entered invalid number')
                log.error('User entered invalid number of cars for renting')
                
            elif self.total_cars < cars :
                print(f'We currently only have {self.total_cars} cars for renting')
                log.error('User want higher number of cars than what are available for renting')
                
            else:
                self.total_cars -= cars
                print(f'Your rented {cars} cars')
                log.info(f'User rented {cars} cars')
                                                           
        except Exception as e:
            log.error(e)    
            

    def return_cars(self, cars):
        try:
            if not isinstance(cars, int):
                print('You entered an invalid number')
                log.error('User entered an invalid number of cars for return')

            else:
                self.total_cars += cars
                print(f'You returned {cars} cars')
                log.info(f'User returned {cars} cars')

        except Exception as e:
            log.error(e)
            
            
    def display_available_cars(self):
        print(f'Total cars available for renting are {self.total_cars}')

vehicle_rental = Vehicle_rent()

vehicle_rental.Available_cars()

vehicle_rental.rent(5)
vehicle_rental.display_available_cars()
vehicle_rental.return_cars(3)
vehicle_rental.display_available_cars()
vehicle_rental.rent('a')
vehicle_rental.rent(10)
```

Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library
, borrow books, and display available books.

```python
import logging as log

log.basicConfig(filename='problem4.log', level=log.DEBUG, format='%(asctime)s- %(message)s - %(levelname)s')


class Library:
    total_books = 55

    def add_books(self, count):
        try:
            if not isinstance(count, int) or count <= 0:
                log.info('User attempted to add an invalid count of books to the library')
                print(f'{count} is not a valid number of books to add in the library')
                log.info(f'User entered an invalid count: {count}')
            else:
                self.total_books += count
                print(f'{count} books added to the library. Total available: {self.total_books}')
                log.info(f'{count} books added to the library')
        except Exception as e:
            log.error(f'Error occurred while adding books: {e}')

    def borrow_books(self, count):
        try:
            if not isinstance(count, int) or count <= 0:
                log.info('User attempted to borrow an invalid count of books from the library')
                print(f'{count} is not a valid number of books to borrow from the library')
                log.info(f'User entered an invalid count: {count}')
            elif self.total_books < count:
                print(f'We currently have only {self.total_books} books in the library')
                log.error('User wanted more books than what is available')
            else:
                self.total_books -= count
                print(f'{count} books borrowed from the library. Total available: {self.total_books}')
                log.info(f'{count} books borrowed from the library')
        except Exception as e:
            log.error(f'Error occurred while borrowing books: {e}')

    def display_available_books(self):
        print(f'Total number of books available in the library: {self.total_books}')

student1 = Library()
student1.display_available_books()
student1.add_books(5)
student1.borrow_books(-5)
student1.borrow_books('a')
student1.borrow_books(50)
```

Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python

class Product:
    def __init__(self, id, name, price, quantity):
        self.id = id
        self.name = name
        self.price = price
        self.quantity = quantity

    def update_quantity(self, new_quantity):
        self.quantity = new_quantity

    def __str__(self):
        return f"ID: {self.id}, Name: {self.name}, Price: {self.price}, Quantity: {self.quantity}"


class Inventory:
    def __init__(self):
        self.products = {}

    def add_product(self, product):
        if product.id not in self.products:
            self.products[product.id] = product
        else:
            print("Product ID already exists.")

    def update_product_quantity(self, product_id, new_quantity):
        if product_id in self.products:
            self.products[product_id].update_quantity(new_quantity)
        else:
            print("Product ID not found.")

    def display_products(self):
        print("Available Products:")
        for product in self.products.values():
            print(product)



inventory = Inventory()

product1 = Product(1, "Keyboard", 20.0, 50)
product2 = Product(2, "Mouse", 10.0, 30)
product3 = Product(3, "Monitor", 150.0, 20)

inventory.add_product(product1)
inventory.add_product(product2)
inventory.add_product(product3)

inventory.display_products()

inventory.update_product_quantity(2, 40)

inventory.display_products()
```

Problem 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.

```python
import logging

logging.basicConfig(filename='problem8.log', level=logging.DEBUG, format='%(asctime)s - %(message)s - %(levelname)s')

class Email:
    def __init__(self, sender, recipient, subject):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.sent = False

    def send_email(self):
        try:
            self.sent = True
            logging.info(f"Email sent from {self.sender} to {self.recipient} with subject: '{self.subject}'")
            print("Email sent successfully!")

        except Exception as e:
            logging.error(f"Error sending email: {str(e)}")
            print(f"Error sending email: {str(e)}")

    def display_email_details(self):
        print(f"Sender: {self.sender}")
        print(f"Recipient: {self.recipient}")
        print(f"Subject: {self.subject}")
        print(f"Email Status: {'Sent' if self.sent else 'Not Sent'}")


# Example Usage
email1 = Email("sender@example.com", "recipient@example.com", "Regarding Meeting")
email1.display_email_details()

# Sending the email
email1.send_email()

# Displaying updated email details
email1.display_email_details()
```

Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

```python
import logging

logging.basicConfig(filename='problem9.log', level=logging.DEBUG, format='%(asctime)s - %(message)s - %(levelname)s')


class social_media:
    
    def __init__(self, username, posts):
        self.username = username 
        self.posts = posts
        
    def new_post(self, add_post):
        log.info(f'User added {add_post} as new post')
        self.posts.append(add_post)
        
    def display_post(self):       
        log.info(f'{self.posts} are the posts')
        print(self.username,self.posts)
        
    def search_post(self, keyword):
        log.info(f'User searched post with {keyword} as a keyword')
        matching = [i for i in self.posts if keyword in i]
        return matching
        
    
s1 = social_media('shiva',['ganpati post', 'diwali post'])

s1.new_post('new year post')

s1.display_post()

s1.search_post('ganpati')
```

Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
import logging

logging.basicConfig(filename='problem10.log', level=logging.DEBUG, format='%(asctime)s - %(message)s - %(levelname)s')


class ToDoList:
    def __init__(self):
        self.tasks = {}  

    def add_task(self, task_name, due_date):
        try:
            
            if task_name in self.tasks:
                raise ValueError("Task already exists")
            
            self.tasks[task_name] = due_date
            logging.info(f"Task '{task_name}' added with due date: {due_date}")
            print(f"Task '{task_name}' added successfully!")

        except ValueError as e:
            logging.error(f"Error adding task: {str(e)}")
            print(f"Error adding task: {str(e)}")

    def mark_task_completed(self, task_name):
        try:
            
            del self.tasks[task_name]
            logging.info(f"Task '{task_name}' marked as completed")
            print(f"Task '{task_name}' marked as completed!")

        except ValueError as e:
            logging.error(f"Error marking task as completed: {str(e)}")
            print(f"Error marking task as completed: {str(e)}")

    def display_pending_tasks(self):
        print("Pending Tasks:")
        for task, due_date in self.tasks.items():
            print(f"Task: {task} | Due Date: {due_date}")


# Example Usage
todo = ToDoList()

todo.add_task("Task 1", "2023-12-05")
todo.add_task("Task 2", "2023-12-10")
todo.add_task("Task 2", "2023-12-20")  
todo.display_pending_tasks()

todo.mark_task_completed("Task 1")

todo.display_pending_tasks()
```
